```python
import easygui as gui
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
filename=gui.fileopenbox()
```

```python
df=pd.read_csv(filename)
df2=df.ix[:,5:10]
```

```python
df.columns
```

```python
df=df[df['You are a '].isin(['JRF', 'Externally Registered PhD Student', 'GS student','Post doc',])]
dfJour=df[(df['Your journey started']!='Left inStem. Not there anymore.') & (df['Your journey started']!='2016')]
```

```python
g = sns.factorplot(hue='Your journey started',y='Is the stipend sufficient?', x="Are you currently being paid?",  data=dfJour,
                   col='You are a ',size=4, aspect=.7,kind='violin',
                   hue_order=['2017', '2015-2016', '2013-2014','2011-2012','Before 2010']);
```

```python
g = sns.factorplot(hue='Your journey started', x="Are you currently being paid?",  data=dfJour,
                   col='You are a ',kind="count",size=4, aspect=.7,
                   hue_order=['2017', '2015-2016', '2013-2014','2011-2012','Before 2010']);
```

```python
sns.countplot(hue='Your journey started', x="Are you currently being paid?",  data=dfJour)
```
