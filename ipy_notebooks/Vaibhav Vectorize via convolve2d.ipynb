```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
n=100
r=np.random.randint(0,5,[n,n])

sc=np.array([[0,1,0],
             [1,0,1],
              [0,1,0]
            ])

#convolve source with sc, pad edges with zeros, output same size as source
con=signal.convolve2d(r,sc,boundary='fill', mode='same')

#plotting fanciness
fig, (ax_orig, ax_mag) = plt.subplots(1, 2, figsize=(10,5))
ax_orig.imshow(r, cmap='gray')
ax_orig.set_title('Original')
ax_orig.set_axis_off()
ax_mag.imshow(np.absolute(con), cmap='gray')
ax_mag.set_title('After neighbour sharing')
ax_mag.set_axis_off()

fig.show()
```

```python
#it is blanzingly fast. About 1ms for a single run!
%timeit con=signal.convolve2d(r,sc,boundary='fill', mode='same')
```
