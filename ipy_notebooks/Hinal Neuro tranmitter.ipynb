```python

import easygui
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib notebook
```

```python
filename = easygui.fileopenbox()
```

```python
df=pd.read_excel(filename)
df.columns=['race','sno','age','chem','norm','ngPerBrain']
df
```

```python
sns.boxplot(y='ngPerBrain',hue='race',x='age',data=df,)
```
