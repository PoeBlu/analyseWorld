```python
linkL=[]
for tr in soup.find('div',attrs={'class':'view-content'}).find('tbody').findAll('tr'):
    url='https://intranet.ncbs.res.in'+tr.find('td',attrs={'class':'views-field views-field-php'}).find('a').attrs['href']
    name=tr.find('td',attrs={'class':'views-field views-field-php'}).find('a').text
    linkL.append([name,url])
    
```
